# Анализ профессий, связанных с данными, в разрезе зарплат. Выбор Топ-5 самых распространенных

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('ds_salaries.csv',encoding='windows-1251',sep=',')
```

```python
df.shape
```

```python
df.head()
```

```python
df.dtypes
```

```python
df=df.rename(columns={'Unnamed: 0':'number'})
```

```python
#количество записей для каждой из профессий
df.groupby('job_title',as_index=False).agg({'number':'count'}).sort_values('number',ascending=False) \
.rename(columns={'number':'total'})
```

```python
df.groupby('job_title',as_index=False).agg({'number':'count'}).sort_values('number',ascending=False) \
.rename(columns={'number':'total'}).head(5)
```

```python
data_analysts=df.query("job_title=='Data Analyst'")
data_analysts.head()
```

# Средняя зарплата дата-аналитика в разрезе стран

```python
#средняя зарплата дата-аналитика в зависимости от страны
company_location_table=data_analysts.groupby('company_location',as_index=False).agg({'salary_in_usd':'mean'}) \
.sort_values('salary_in_usd',ascending=False).rename(columns={'salary_in_usd':'average_salary_in_usd'})

company_location_table
```

```python
sns.barplot(x='company_location',y='average_salary_in_usd', data=company_location_table)
```

# Средняя зарплата дата-аналитика в разрезе размера компании

```python
#средняя зарплата дата-аналитика в зависимости от размера компании
company_size_table=data_analysts.groupby('company_size',as_index=False).agg({'salary_in_usd':'mean'}) \
.sort_values('salary_in_usd',ascending=False).rename(columns={'salary_in_usd':'average_salary_in_usd'})

company_size_table
```

```python
sns.barplot(x='company_size',y='average_salary_in_usd', data=company_size_table)
```

# Максимальная зарплата дата-аналитика

```python
max_salary_in_usd=data_analysts.salary_in_usd.max()
```

```python
data_analysts.query('salary_in_usd==@max_salary_in_usd')
```

# Сравнение средних зарплат топ-5 профессий в зависимости от experience_level

```python
top_5_job=df.query("job_title=='Data Scientist' or job_title=='Data Engineer' or job_title=='Data Analyst' or job_title=='Machine Learning Engineer' or job_title=='Research Scientist'")

top_5_job.head()
```

```python
#сравнение средних зарплат топ-5 профессий в зависимости от experience_level
top_5_job_table=top_5_job.groupby(['job_title','experience_level'],as_index=False).agg({'salary_in_usd':'mean'}) \
.sort_values('salary_in_usd',ascending=False).rename(columns={'salary_in_usd':'average_salary_in_usd'})

top_5_job_table
```

```python
plt.figure(figsize=(15,15))
sns.barplot(x='job_title',y='average_salary_in_usd', data=top_5_job_table)
```

# Расчет премии в зависимости от experience_level

```python
#расчет премии в зависимости от experience_level и добавление в датафрейм data_analysts
premium_se_ex=data_analysts.query("experience_level=='SE' or experience_level=='EX'")
premium_se_ex['premium']=data_analysts.salary_in_usd*0.1
```

```python
premium_se_ex.head()
```

```python
premium_en_mi=data_analysts.query("experience_level=='EN' or experience_level=='MI'")
premium_en_mi['premium']=data_analysts.salary_in_usd*0.05
```

```python
premium_en_mi.head()
```

```python
data_analysts=premium_se_ex.append(premium_en_mi)
```

```python
data_analysts.head()
```

# Максимальная и минимальная зарплата в разрезе experience_level и страны

```python
max_min_salary=data_analysts.groupby(['experience_level','company_location'],as_index=False)['salary_in_usd'] \
.agg(['max','min'])
max_min_salary
```

Выводы: Топ-5 самых распространенных профессий, связанных с данными, - Data Scientist, Data Engineer, Data Analyst, Machine Learning Engineer, Research Scientist.

Максимальная средняя зарплата в US - 107674, минимальная в IN - 6072.

Максимальная зарплата Data Analyst у сотрудника с позицией SE в US - 200000 usd. Размер компании L.
